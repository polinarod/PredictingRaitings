```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import sklearn
from sklearn.feature_selection import VarianceThreshold
```

```python
features=pd.read_csv(r'C:\Users\Asus\PredictingRatings\data\features.csv')
#features=pd.read_csv(r'C:\Users\vznam\Downloads\PredictingRatings-master\data\features.csv')

print(features.info())
print(features.shape)
features.describe()
```

```python
cols=features.columns
for c in cols:
    if c.find('End') != -1:
        print(c)
```

```python
from sklearn.feature_selection import VarianceThreshold

sel = VarianceThreshold(threshold=(.8 * (1 - .8)))
sel_fit = sel.fit(features)
transform=sel_fit.transform(features)
features_new=pd.DataFrame(transform)
features_new.shape
```

```python
inv_trans=sel_fit.inverse_transform(features_new)
inv_features=pd.DataFrame(inv_trans)
reject_features=inv_features.columns[(inv_features== 0).all()]
reject_features
```

```python
rej=[]
for ind,val in enumerate(reject_features):
    rej.append(features.columns[val])
    print(features.columns[val])
```

```python
reject_columns=pd.Series(rej)
reject_columns
```

```python
reject_columns.to_csv(r'C:\Users\Asus\PredictingRatings\data\reject_columns.csv',sep=',',index=False)
```

```python
features=pd.read_csv(r'C:\Users\Asus\PredictingRatings\data\features_new.csv')
features.head()
```

```python
corr=features.corr(method='pearson')
corr.head()
```

```python
indexes=corr.index
corr_feat=[]
corr_index=[]
corr_columns=[]
for index in corr.index:
    for column in corr.index:
        if index < column and abs(corr.loc[index,column])>=0.9:
            corr_feat.append((index,column))
            corr_index.append(index)
            corr_columns.append(column)
            
print(len(corr_feat))
corr_feat
```

```python
features_corr=features.drop(columns=corr_columns)
features_corr.shape
```

```python
print(corr_columns)
len(corr_columns)
```

```python
corr1=features_corr.corr(method='pearson')
corr1.head()
```

```python
corr_feat1=[]
corr_index1=[]
corr_columns1=[]
for index in corr1.index:
    for column in corr1.columns:
        if index < column and abs(corr1.loc[index,column])>=0.8:
            corr_feat1.append((index,column))
            corr_index1.append(index)
            corr_columns1.append(column)
            
print(len(corr_feat1))
corr_feat1
```

```python
features_corr1=features_corr.drop(columns=corr_columns1)
features_corr1.shape
```

```python
cols=list(set(corr_columns+corr_columns1))
```

```python
len(cols)
```

```python
reject_cols=pd.Series(cols)
reject_cols.to_csv(r'C:\Users\Asus\PredictingRatings\data\reject_cols.csv',sep=',',index=False)
```
